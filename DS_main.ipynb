```python
# отделяем предикторы от зависимой переменной
features = df.drop(['highest_yearly_earnings', 'category', 'country', 'channel_type', 'created_month', 'created_date'], axis=1)
target = df['highest_yearly_earnings']

# разбиваем на тестовую и обучающую подвыборки
features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size=0.2, random_state=42)

print('Размер обучающей выборки с признаками:', features_train.shape[0])
print('Размер обучающей выборки с у:',target_train.shape[0])
print('Размер тестовой выборки с признаками:',features_test.shape[0])
print('Размер тестовой выборки с у:',target_test.shape[0])
```

```python
# подключение Keras
from tensorflow import keras

# создание модели: инициализируем нейронную сеть. Модели задаётся класс Sequential (последовательный). Этот класс применяется 
# для моделей, в которых слои идут последовательно. Слой — набор нейронов с общим входом и выходом.
model = keras.models.Sequential()
# Сеть будет состоять из одного нейрона, или значения на одном выходе. В ней n входов (=предикторов), каждый умножается на свой 
# вес. Например, x₁ умножается на w₁. Есть ещё один вход, который всегда равен единице. Его вес обозначается b (смещение). 
# Именно из подбора весов w и b состоит процесс обучения нейронной сети. После того как все произведения значений входов на 
# веса просуммированы, на выход подаётся ответ нейронной сети (а).

# параметры нейронной сети. Команда keras.layers.Dense() - создаёт один слой нейронов. Dense - каждый вход будет соединён с 
# каждым нейроном, или выходом. Параметр units - задаёт количество нейронов в слое, input_dim — количество входов в слое. 
# Причём в этом параметре смещение не учитывается. keras.layers - это так создаем слой для сети + тут количество входов 
# берётся из обучающей выборки 
# Чтобы к модели добавить созданный слой, вызовите метод model.add() (англ. «добавить в модель»):
model.add(keras.layers.Dense(units=1, input_dim=features.shape[1]))

# как обучается нейронная сеть. Эта строчка готовит модель к обучению. После этой команды конструкцию сети уже нельзя будет 
# изменить. В параметре loss указана функция потерь для задачи регрессии — MSE. В параметре optimizer='sgd' задан метод 
# градиентного спуска. Запомните: нейронные сети обучаются SGD.
# model.compile(loss='mean_squared_error', optimizer='sgd')
model.compile(loss='mean_absolute_percentage_error', optimizer='sgd')

# обучить модель. Теперь можно запустить обучение модели:
model.fit(features, target, verbose=2, epochs = 5)
# Чтобы формат ответа был понятным, нужен аргумент verbose=2, где 2 - вывод в консоль. 0 -вывода не будет вовсе; 
# 1 - вывод предназначен для Jupyter Notebook. Значение MSE уменьшается с каждой эпохой. Когда вы обучаете нейронную сеть, 
# обычно запускается много эпох. Остановите обучение, если достигли нужного качества модели или когда метрика перестала 
# снижаться.
# loss - среднее значение функции потерь на первой эпохе.
```

### Многослойные сети

```python
# Проблема: С увеличением количества слоёв обучение ухудшается. Чем больше слоёв в сети, тем меньше сигнала от входа доходит 
# до выхода сети. Это называется vanishing signal. Причина затухания кроется в сигмоиде, преобразующей большие значения в 
# маленькие по много раз. Чтобы избавиться от затухания сигнала, можно выбрать другую функцию активации. Например, ReLU 
# (Rectified Linear Unit). ReLU приводит к нулю всё, что меньше 0, и пропускает положительные значения без изменений.
```

```python
# Слои добавляются методом model.add(). Аргумент input_dim нужно указать только для первого слоя.
# Добавим в нейронную сеть ещё 1 слой. Пусть у первого скрытого слоя будет 10 нейронов units с сигмоидной активацией. 
# Во втором выходном слое будет 1 нейрон с сигмоидой: её рассмотрим как вероятность класса «1». Обучим нейронную сеть на 
# 10 эпохах, напечатав на экране прогресс обучения.

model = keras.models.Sequential()
model.add(keras.layers.Dense(units = 10, input_dim = features_train.shape[1], 
                             activation='sigmoid'))
model.add(keras.layers.Dense(units = 1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['acc'])

model.fit(features_train, target_train, epochs=10, verbose=2)
```
